# Method Chaining

````python
import numpy as np
import pandas as pd
import seaborn as sns
````

````python
iris = sns.load_dataset('iris')
````

````python
iris
````

````python
iris.index.name = 'iris_id'
````

````python
iris
````

````python
iris = iris.reset_index()\
    .set_index(['species', 'iris_id'])
````

````python
iris
````

````python
iris.groupby('species')\ 
    .agg({'mean', 'min', 'max'})\
    .T\
    .style.background_gradient(cmap="GnBu", axis=1)
````

````python
iris
````

````python
# iris.T
````

````python
X = iris.T.stack()
````

````python
X
````

````python
X.index.names = ['feature', 'iris_id']
````

````python
X
````

````python
X.groupby('feature')\
    .agg({'mean', 'max', 'min'})\
    .T\
    .style.background_gradient(cmap="GnBu", axis=0)
````
